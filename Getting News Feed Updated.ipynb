```python
from GoogleNews import GoogleNews # importing the google news package to fetch the news
import pandas as pd
import numpy as np
```

```python
googlenews = GoogleNews() 
```

```python
# Setting up the basic parameters; Date range can be changed accordingly
googlenews.set_lang('en')
googlenews = GoogleNews(period='1d')
googlenews.set_encode('utf-8') 
```

```python
df=pd.DataFrame(columns=['Date','Time','Title','Desc','Link']) # Dataframe to store the news
```

```python
# results are passed as dictionary, So using the Looping method to save the headlines on the retrieved pages


def newsfeed1(df,raw_dictionary): # Gathering all the data of the current page to one dataframe
    for i in range(len(raw_dictionary)-1):
        if raw_dictionary is not None:
            d=raw_dictionary[i]['datetime']
            date=pd.to_datetime(d)
            title=raw_dictionary[i]['title']
            time=raw_dictionary[i]['date']
            desc=raw_dictionary[i]['desc']
            link=raw_dictionary[i]['link']
            df = df.append({'Date': date,'Time':time, 'Title': title,
                        'Desc': desc,'Link':link}, ignore_index=True)
        else:
            break
        
    return df
```

```python
keywords=['Apple-Stock','Apple-Revenue','Apple-Sales','Apple','AAPL'] # List of the Keyswords: Which we want to search it can be anything like "Business, Stocks, Apple, Economy"
googlenews.set_time_range('12/01/2020','03/10/2021') #Format: MM/DD/YYYY  # increase the time range to get more results 

final_df=pd.DataFrame() # Final dataframe containing the news of all the keywords Searched.

for steps in range(len(keywords)):
    string=(keywords[steps])
    googlenews.search(string)    # Each Keyword will be searched seperately and resutls will be saved in One Dataframe only
    t=googlenews.results() # Fetching the results and saving them offline
    total_pages=10
    
    
    for steps in range(total_pages):
        googlenews.get_page(steps) # offline variable consists of pages specified by user so using "for loop" to retrieve all the data in dataframe
        feed=newsfeed1(df,t)
    
    final_df=final_df.append(feed)
    googlenews.clear()         # clearing off the search results of previous Keyword to avoid duplication
    
shape=final_df.shape[0]   
final_df.index=np.arange(shape) # Resetting the index of the final result
final_df
```
